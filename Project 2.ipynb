```python
nex=df.groupby(['Fiscal Year'],as_index=False)['Apps Received'].sum().sort_values(by='Apps Received',ascending=False) 
sns.barplot(x='Fiscal Year',y='Apps Received',data=nex)
```

```python
The above graph shows the decline in Job Applications in between the Fiscal Years 2013-2014 and 2014-2015.
```

```python
new=pd.DataFrame(df,columns=['Female','Male','Unknown_Gender','Job Number'])
df.plot(x='Job Number',y=['Female','Male','Unknown_Gender'],kind='bar',figsize=(40,40))
plt.show()
```

```python
The above graph shows the number of applications recieved for each Job opening by different genders.
```
